```python
import seaborn as sns
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
plt.style.use('classic')
%matplotlib inline
```

<h1>matplotlib plotting</h1>

```python
# Create some data
rng = np.random.RandomState(0)
x = np.linspace(0, 10, 500)
y = np.cumsum(rng.randn(500, 6), 0)
```

```python
plt.plot(x,y)
plt.legend('ABCDEF', ncol=2, loc='upper left')
```

<h1>Seaborn plotting</h1>

```python
sns.set()
data = np.random.multivariate_normal([0, 0], [[5, 2], [2, 2]], size=2000)
data = pd.DataFrame(data, columns=['x', 'y'])
```

<h1>Histogram and density plotting</h1>

```python
for col in 'xy':
    plt.hist(data[col], density=True, alpha=0.5)
```

<h1>Kernel Density Estimation(KDE) for smooth distribution estimation</h1>

```python
for col in 'xy':
    sns.kdeplot(data[col], shade=True)
```

```python
# Combined KDE and Histogram
sns.distplot(data['x'])
sns.distplot(data['y'])
```

```python
# Joint distro and Marginal distro
with sns.axes_style('white'):
    sns.jointplot("x", "y", data, kind='kde');
```

```python
# Hexagonal histogram as parameter
with sns.axes_style('white'):
    sns.jointplot("x", "y", data, kind='hex');
```

<h1>Pair plotting</h1>

```python
# import iris dataset
iris = sns.load_dataset("iris")
iris.head()
```

```python
sns.pairplot(iris, hue='species', height=2.5)
```

<h1>Faceted Histogram</h1>

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips['tip_pct'] = 100 * tips['tip'] / tips['total_bill']
grid = sns.FacetGrid(tips, row="sex", col="time", margin_titles=True)
grid.map(plt.hist, "tip_pct", bins=np.linspace(0, 40, 15));
```

<h1>Factor plot</h1>

```python
with sns.axes_style(style='ticks'):
    g = sns.catplot("day", "total_bill", "sex", data=tips, kind="box")
    g.set_axis_labels("Day", "Total Bill");
```

<h1>Joint Distribution</h1>

```python
with sns.axes_style('white'):
    sns.jointplot("total_bill", "tip", data=tips, kind='hex')
```

<h1>Regression and Decision</h1>

```python
# The joint plot can even do some auto kernel density estimation and regression
sns.jointplot("total_bill", "tip", data=tips, kind='reg');
```

<h1>Bar Plot</h1>

```python
planets = sns.load_dataset('planets')
planets.head()
```

```python
with sns.axes_style('white'):
    g = sns.catplot("year", data=planets, kind="count")
    g.set_xticklabels(step=5)
```

```python
with sns.axes_style('white'):
    g = sns.catplot("year", data=planets, aspect=2.0, kind='count', hue='method')
    g.set_ylabels('Number of Planets Discovered')
    g.set_xticklabels(step=2)
```
